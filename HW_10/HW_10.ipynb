## Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
%%time

ratings = pd.read_csv('ml-25m/ratings.csv')
movies = pd.read_csv('ml-25m/movies.csv')
```

```python
%%time

join = ratings.merge(movies, on='movieId', how='right')
```

```python
top_movie = join[join['rating'] == 5.0]
top_movie_sorted = top_movie['title'].value_counts().sort_values(ascending = False)

print(f'Фильм с большим количеством оценки 5.0 {top_movie_sorted.keys()[0]}')
```

```python
print(top_movie_sorted[:5])
```

## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data_baltic = pd.read_csv('power.csv')
data_baltic.head()
```

```python
data_baltic = data_baltic[(data_baltic['country']=='Latvia') \
                          | (data_baltic['country']=='Lithuania') \
                          | (data_baltic['country']=='Estonia') ]

data_baltic = data_baltic[(data_baltic['category']==4) \
                          | (data_baltic['category']==12) \
                          | (data_baltic['category']==21)]

data_baltic = data_baltic[(data_baltic['year'] >= 2005) | (data_baltic['year'] <= 2010)]

data_baltic = data_baltic[(data_baltic['quantity'] >= 0)]

total_quantity = sum(data_baltic['quantity'])
```

```python
print(total_quantity)
```
